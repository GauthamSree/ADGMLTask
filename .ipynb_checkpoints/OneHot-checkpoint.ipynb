```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats

%matplotlib inline
```

```python
data = pd.read_csv("preOneHotdataOnly.csv")
data.head()
```

```python
""" separating features and target """
Y = data['Infect_Prob']
X = data.drop(['Infect_Prob'],axis=1)
```

```python
from sklearn.model_selection import train_test_split

train_X,test_X,train_Y,test_Y = train_test_split(X,Y)

print("Training Size : ",train_X.shape)
print("Testing Size : ",test_X.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
import math


regressor = RandomForestRegressor()
regressor.fit(train_X,train_Y)
    
def rmse(x,y): return math.sqrt(((x-y)**2).mean())
def print_score(m):
    res = [m.score(test_X, test_Y)]
    print(res)
y_pred = regressor.predict(test_X)
print_score(regressor)
print("Root Mean Square Error = ", rmse(y_pred,test_Y))
print(rmse(test_Y, y_pred))
```

```python
features = train_X.columns
importances = regressor.feature_importances_
std = np.std([tree.feature_importances_ for tree in regressor.estimators_],
             axis=0)
indices = np.argsort(importances)[::-1]

# Print the feature ranking
print("Feature ranking:")

for f in range(train_X.shape[1]):
    print("%d. feature %d (%f)" % (f + 1, indices[f], importances[indices[f]]))

# Plot the feature importances of the forest

plt.figure(figsize=(20,10))
plt.title("Feature importances")
plt.bar(train_X.columns, importances[indices],
       color="r", yerr=std[indices], align="center")
plt.xticks(range(train_X.shape[1]), features[indices], rotation = 'vertical')
plt.xlim([-1, train_X.shape[1]])
```
